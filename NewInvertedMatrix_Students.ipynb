# What Is an Inverted Index and When to Use It?

An inverted index is a data structure used to map words (terms) to the documents (or positions within documents) that contain them.
It’s called “inverted” because it flips the normal relationship, instead of listing words in a document, it lists documents for each word.

TERMS --> DOCUMENTS

| Term | Documents |
| ---- | --------- |
| cat  | D1, D3    |
| dog  | D1, D2    |
| eats | D2, D3    |

This structure powers most search engines, allowing quick retrieval of all documents containing a term.


### When to Use an Inverted Index

Use an inverted index when you need fast full-text search or information retrieval tasks such as:

* Search engines (Google, Lucene)

* Document similarity analysis

* Keyword-based filtering or recommender systems

It’s optimal when:

* You have large collections of text documents

* You need to find documents matching keywords quickly

* You can preprocess data (build the index once, query many times)

## Variants and Related Models

### Biword (Bigram) Indexes

A biword index stores pairs of consecutive words as entries.

Example:
* "new york city is big"

Biword index entries:
| Term (biword) | Documents |
| ------------- | --------- |
| new_york      | D1        |
| york_city     | D1        |
| city_is       | D1        |
| is_big        | D1        |

Use case:

* Improves phrase search accuracy ("new york" as a phrase)

* Used when you want to support exact phrase matching efficiently

Trade-offs:

* Larger index (more entries)

* Limited to fixed phrase lengths (usually 2-grams)

```python
text = "new york city is big"
words = text.lower().split()

biword_index = defaultdict(list)

for i in range(len(words) - 1):
    biword = words[i] + " " + words[i + 1]
    biword_index[biword].append(0)

print(dict(biword_index))
```

### Positional Indexes

A positional index stores positions of terms in each document.

Example:

| Term | Documents (positions) |
| ---- | --------------------- |
| new  | D1: [1], D2: [4]      |
| york | D1: [2], D2: [5]      |
| city | D1: [3], D3: [7]      |

Now you can check if “new” and “york” occur adjacently (positions 1 and 2 → "new york").

Use case:

* Needed for phrase queries, proximity search

* Enables queries like “find ‘AI model’ within 3 words of ‘language’”

Trade-offs:

* More storage required (because of position lists)

* More complex query processing

```python
text = "the cat sat on the mat"
words = text.lower().split()

positional_index = defaultdict(list)

for pos, word in enumerate(words):
    positional_index[word].append((0, pos))

print(dict(positional_index))
```

## Term-Document Count Matrix (Document-Term Matrix)

This is a matrix representation (used in ML/NLP) where:

* Rows = documents

* Columns = terms

* Entries = frequency (count) of term in document

Example:

| Document | cat | dog | eats |
| -------- | --- | --- | ---- |
| D1       | 1   | 1   | 0    |
| D2       | 0   | 1   | 1    |
| D3       | 1   | 0   | 1    |

Use case:

* Statistical and machine learning models (text classification, clustering, topic modeling)

* When you need a numerical feature representation of text

Trade-offs:

* Very large and sparse (many zeros)

* Slower for direct lookup; better for ML rather than keyword search

```python
from collections import Counter

documents = [
    "cats and dogs",
    "dogs are cute",
    "cats are cute"
]

tokenized_docs = [doc.lower().split() for doc in documents]
vocab = sorted(set(word for doc in tokenized_docs for word in doc))

matrix = []

for doc in tokenized_docs:
    counts = Counter(doc)
    row = [counts[word] for word in vocab]
    matrix.append(row)

print("Vocabulary:", vocab)
print("Matrix:")
for row in matrix:
    print(row)
```

## Bag of Words (BoW) Model

A conceptual model, not a data structure.
It represents text as an unordered collection of words, ignoring grammar and word order.

Example for “new york city” → {new:1, york:1, city:1}

Use case:

* Text classification, sentiment analysis, spam detection

* Simple yet effective baseline representation for machine learning

Trade-offs:

* Loses context and word order

* Cannot handle phrases or proximity queries

* Often combined with weighting (like TF-IDF)


| Concept                         | Output Type                      | Example                             |
| ------------------------------- | -------------------------------- | ----------------------------------- |
| **Bag of Words (BoW)**          | Dictionary per document          | `{'cats': 1, 'dogs': 1, 'love': 1}` |
| **Term–Document Matrix (TDCM)** | Matrix / table for all documents | `[[1,1,1,1,1], [0,0,1,0,0], ...]`   |




```python
from collections import Counter

text = "new york city new city"
words = text.lower().split()

bow = Counter(words)

print(dict(bow))
```

| Technique            | Stores                  | Preserves Word Order? | Main Use                  | Pros                   | Cons                      |
| -------------------- | ----------------------- | --------------------- | ------------------------- | ---------------------- | ------------------------- |
| **Inverted Index**   | Term → Docs             | No                     | Keyword search            | Fast lookups           | No phrase info            |
| **Biword Index**     | 2-word phrases → Docs   | Yes (2-word only)       | Phrase search             | Handles common phrases | Larger index              |
| **Positional Index** | Term → (Doc, Positions) | Yes                     | Proximity / Phrase search | Flexible, accurate     | High storage              |
| **Term-Doc Matrix**  | Doc × Term counts       | No                     | ML features               | Numeric, easy for ML   | Sparse, not for retrieval |
| **BoW Model**        | Term frequencies        | No                     | ML / NLP                  | Simple, effective      | Loses order/context       |

```python
import nltk
nltk.download('movie_reviews')

from nltk.corpus import movie_reviews

documents = [movie_reviews.raw(fileid) for fileid in movie_reviews.fileids()[:10]]  # first 10 docs
print(documents[0][:200])
```

```python
import re
from nltk.corpus import stopwords

# Download stopwords the first time
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess(text):
    # Lowercase
    text = text.lower()
    # Keep only words
    words = re.findall(r'\b[a-z]+\b', text)
    # Remove stopwords
    words = [word for word in words if word not in stop_words]
    return words
```

```python
from collections import defaultdict

def build_inverted_index(docs, preprocess_fn=None):
    """
    Build an inverted index from a list of documents.
    
    Arguments:
        docs (list of str): List of document texts.
        preprocess_fn (function): Function to preprocess each document (optional).
    
    Returns:
        dict: inverted index {term: [doc_ids]}
    """
    index = defaultdict(list)
    
    for doc_id, doc in enumerate(docs):
        # Apply preprocessing if provided
        if preprocess_fn:
            words = preprocess_fn(doc)
        else:
            words = doc.lower().split()
        
        # Add doc_id to each unique word
        for word in set(words):
            ## add here your transformations
            index[word].append(doc_id)
            
    return dict(index)
```

```python
# Build inverted index with preprocessing
inverted_index = build_inverted_index(documents, preprocess_fn=preprocess)

# Show example
term_to_search = "girlfriend"
#print your results here
print(term_to_search, "->", inverted_index.get(term_to_search, "Word not found"))

term_to_search = "game"
#print your results here
print(term_to_search, "->", inverted_index.get(term_to_search, "Word not found"))

```
