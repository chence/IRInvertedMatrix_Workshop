
# 🎓 Student Instructions  
### NLP Workshop Submission Guide

Welcome to the NLP Workshop! This notebook explains how to complete your assignment and submit it automatically.

---

## 1. **Setup Your Repository**
1. Fork the instructor’s repository on GitHub:
   - Go to: [Instructor Repo](https://github.com/ProfEspinosaAIML/IR_Inverted_Matrix_Workshop.git)  
   - Click **Fork** to create your own copy.

2. Clone your fork:
   ```bash
   git clone https://github.com/<your_username>/IR_Inverted_Matrix_Workshop.git
   cd IR_Inverted_Matrix_Workshop
   ```

3. Install required dependencies:
   ```bash
   pip install -r requirements.txt
   ```

---

## 2. **Complete the Workshop**
- Open the notebook:
  ```bash
  jupyter notebook IR_InvertedMatrix_Workshop.ipynb
  ```
- Fill in all the sections:  
  - Replace **TODO** placeholders in **Markdown cells** with your answers.  
  - Implement the required functions in **Code cells** (e.g., `build_inverted_index`, `query_processor`).  
  - Run all code cells to check your solutions.  

---

## 3. **Submit Your Work**
1. At the end of the notebook, run this cell:
   ```python
   !python submit_assignment.py --notebook IR_InvertedMatrix_Workshop.ipynb --student_id "your_team_id"
   ```

   Replace `"your_team_id"` with your actual team name or ID (e.g., `"team1"`).

2. What happens:
   - ✅ The script checks that all sections are completed.  
   - 🧪 Required functions are tested with sample inputs.  
   - 📊 A **Validation Report** is shown with:  
     - Errors ❌ if something is missing or incorrect.  
     - Warnings ⚠️ if some cells were not executed.  
     - Score (points earned vs total).  
   - 💾 The results are logged in `submissions_log.csv`.  
   - 🚀 If validation passes, your notebook is committed and pushed to a branch:  
     ```
     submissions/your_team_id
     ```

3. Example output:
   ```
   === Validation Report ===
   ✅ All required checks passed.

   Score: 10 / 10 points
   ✅ Submission pushed to branch: submissions/team1
   ```

---

## 4. **Fixing Errors**
- If your submission fails, the validator will tell you what went wrong (e.g., missing function, unfilled markdown, wrong return type).  
- Fix the issues, re-run all cells, and submit again.  
- Only the **latest submission** will remain on your team’s branch.  

---

## 5. **Checklist Before Submitting**
- [ ] All Markdown sections are filled (no `TODO` placeholders).  
- [ ] Required functions implemented correctly.  
- [ ] All code cells executed successfully.  
- [ ] Notebook runs from top to bottom without errors.  

---

## 6. **After Submission**
- Your instructor will review your branch.  
- Scores are automatically recorded in `submissions_log.csv`.  
- You can resubmit until the deadline.  

---

✅ That’s it! You now know how to complete and submit your NLP workshop assignment.
