
# 📘 Instructor Instruction Set  
### NLP Workshop with Auto-Submission + Auto-Grading

---

## 1. **Prepare the Repository**
1. Clone the master repo:
   ```bash
   git clone https://github.com/ProfEspinosaAIML/IR_Inverted_Matrix_Workshop.git
   cd IR_Inverted_Matrix_Workshop
   ```
2. Ensure the following structure exists:
   ```txt
   IR_Inverted_Matrix_Workshop/
   ├── IR_InvertedMatrix_Workshop.ipynb    # Student workshop notebook
   ├── submit_assignment.py                 # Submission + push script
   ├── requirements.txt                     # Dependencies (nbformat, gitpython)
   ├── README.md                            # Student instructions
   ├── config/
   │   └── required_items.json              # Instructor-editable requirements
   └── utils/
       └── validate_notebook.py             # Validator + scoring + logging
   ```
3. Edit `config/required_items.json` to set:  
   - **Functions students must implement**, with test cases and points.  
   - **Markdown sections students must fill**, with points.  

   Example:
   ```json
   {
     "required_functions": {
       "build_inverted_index": {
         "test_input": [["this is a doc", "this doc is about nlp"]],
         "expected_type": "dict",
         "points": 5
       },
       "query_processor": {
         "test_input": ["doc", {"doc1": [0], "doc2": [1]}],
         "expected_type": "list",
         "points": 5
       }
     },
     "required_markdown": {
       "Introduction": 2,
       "Reflection": 3
     }
   }
   ```

---

## 2. **Scaffold Students**
1. Provide students with the GitHub repo link.  
   Students should **fork and clone**:
   ```bash
   git clone https://github.com/<student_team>/IR_Inverted_Matrix_Workshop.git
   ```
2. Ask them to install dependencies:
   ```bash
   pip install -r requirements.txt
   ```
   Contents of `requirements.txt`:
   ```
   nbformat
   gitpython
   ```
3. Explain the workflow:
   - Students edit **`IR_InvertedMatrix_Workshop.ipynb`**.  
   - Notebook contains:
     - Learning objectives  
     - Scaffolded tasks (code + markdown placeholders)  
     - Instructor notes  
     - Submission checklist  

---

## 3. **Student Submission Workflow**
1. In the last cell of the notebook, students run:
   ```python
   !python submit_assignment.py --notebook IR_InvertedMatrix_Workshop.ipynb --student_id "team1"
   ```
2. The script will:
   - Validate the notebook:  
     - All required Markdown sections filled.  
     - Required functions exist.  
     - Functions pass quick tests.  
   - Generate a **validation report** with:  
     - ✅/❌ per requirement  
     - Warnings for unexecuted cells  
     - Score breakdown (earned vs total points)  
   - Log the result into `submissions_log.csv`.  
   - If all checks pass → commit & push to branch `submissions/<student_id>`.  
   - If checks fail → submission blocked, but attempt logged.  

---

## 4. **Instructor Grading & Collection**
1. Pull all student submission branches:
   ```bash
   git fetch --all
   ```
2. Inspect specific submissions:
   ```bash
   git checkout submissions/team1
   git checkout submissions/team2
   ```
   Each branch contains only the student’s **latest notebook submission**.  
3. Review auto-generated gradebook:
   ```bash
   cat submissions_log.csv
   ```
   Example:
   ```csv
   timestamp,student_id,notebook,score,max_score,status
   2025-09-08T14:22:01,team1,IR_InvertedMatrix_Workshop.ipynb,8,10,✅ Passed
   2025-09-08T14:25:12,team2,IR_InvertedMatrix_Workshop.ipynb,6,10,❌ Failed
   ```
4. Optionally, import `submissions_log.csv` into Excel or a Learning Management System (LMS) for record keeping.  

---

## 5. **Instructor Best Practices**
- 🔑 **Before release:**  
  - Update `required_items.json` to match assignment learning objectives.  
  - Test validator on a sample notebook.  
- 📢 **During workshops:**  
  - Remind students to **execute all cells** before submission.  
- 📊 **After deadline:**  
  - Use `submissions_log.csv` for quick scoring.  
  - Optionally review student notebooks in branches for qualitative feedback.  
- 🔒 **Security:**  
  - Ensure students set up **GitHub SSH keys or PATs**.  
  - Recommend SSH for smoother automation.

- Get **auto-graded submissions** logged in real time.  
- Collect everything into GitHub branches + CSV gradebook.
